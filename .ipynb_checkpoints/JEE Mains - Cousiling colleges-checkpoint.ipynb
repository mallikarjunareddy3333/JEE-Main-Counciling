## All Institutes

```python
print(len(jee_main_df['Institute'].unique()))
jee_main_df['Institute'].unique()
```

## Mapping Program with branch_code

```python
programs_df = pd.read_excel('all_branch.xlsx')
programs_df.head(20)
```

```python
jee_main_df = pd.merge(jee_main_df, programs_df, on='Program')
jee_main_df.head()
```

## Grouping the colleges

```python
def group_college(x):
    if 'National Institute of Technology'.lower() in x.lower():
        return "NIT"
    elif 'Indian Institute of Information Technology'.lower() in x.lower():
        return "IIIT"
    elif 'International Institute of Information Technology'.lower() in x.lower():
        return "IIIT"
    else:
        return "GFTI"
```

```python
jee_main_df['college_type'] = jee_main_df['Institute'].apply(group_college)
```

```python
jee_main_df['Institute'].loc[jee_main_df['college_type'] == 'IIIT'].unique()
```

## Generate excel file with 'Institute & college_type' to give ranking

```python
all_colleges = pd.DataFrame(jee_main_df.groupby('Institute')['college_type'].apply(lambda x: list(np.unique(x))[0]))
all_colleges.reset_index()

print(all_colleges.head())

all_colleges.to_excel('all_colleges_new.xlsx')
```
