```python
import numpy as np
import pandas as pd
```

```python
def loadAndCombineAllPagesOfSheet(sheet_name, limit):
    csab_round1 = pd.read_excel(sheet_name, sheet_name = None)
    column_names = ["instcd","Institute","Brcd","Program","Quota","Category","Gender","Opening","Closing"]
    csab_round = pd.DataFrame(columns = column_names)
    
    for i in range(1,limit):
        temp = pd.DataFrame(csab_round1['Page '+ str(i)]).iloc[2:, :9]
        temp.columns = column_names
        csab_round = pd.concat([csab_round, temp ])
    
    csab_round = csab_round.reset_index()
    csab_round = csab_round[column_names]
    return csab_round
```

```python
csab_round1 = loadAndCombineAllPagesOfSheet('csab/Round1.xlsx', 173)
csab_round2 = loadAndCombineAllPagesOfSheet('csab/Round2.xlsx', 65)
```
